### Example mapping functions

```python
import numpy as np
import pandas as pd
import xarray as xr

import matplotlib
import matplotlib.pyplot as plt

import cartopy.crs as ccrs
import cartopy.feature as cfeature
```

```python
import sys
sys.path.append('~/leap-carbon-ppe/')
from linnia_functions import *
```

### load data

```python
# load dataset
filename = '/glade/derecho/scratch/linnia/leap-reu-mf-2025/ctsm6ppe_amean_1850-2023.nc'
dataset = xr.open_dataset(filename)
ds = dataset.isel(ens =slice(1,1494))
```

```python
# select ET for the default ensemble member and average over 2000-2010
data = ds.EFLX_LH_TOT_gridded_amean.sel(ens=0,year=slice('2000','2010')).mean(dim='year')
```

### Mapping

```python
def get_map(da,sgmap=None):
    if not sgmap:
        sgmap=xr.open_dataset('/glade/u/home/linnia/ctsm6_ppe/utils/sgmap_retrain_h0.nc')
    return da.sel(gridcell=sgmap.cclass).where(sgmap.notnan).compute()
```

```python
mapped_data = get_map(data)
```

```python
fig, ax = plt.subplots(1,1,figsize=(12, 6), subplot_kw={'projection': ccrs.Robinson()})
plt.rcParams['font.size'] = 16

ax.coastlines()
ocean = ax.add_feature(cfeature.NaturalEarthFeature('physical', 'ocean', '110m',facecolor='white'))
pcm = ax.pcolormesh(mapped_data.lon, mapped_data.lat, mapped_data,
                    transform=ccrs.PlateCarree(), shading='auto',
                    cmap='BrBG', 
                    vmin=0,
                    vmax=150)

cbar = fig.colorbar(pcm, ax=ax, fraction=0.08,orientation='horizontal')
cbar.set_label('Evapotranspiration (W/m2)', size=16)
ax.set_title('Default CLM ET \n (2000-2010)')

```
