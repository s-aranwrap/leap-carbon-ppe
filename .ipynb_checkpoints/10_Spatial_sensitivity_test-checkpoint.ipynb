```python
# Load dataset

import xarray as xr
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, r2_score
import pickle
from sklearn.preprocessing import StandardScaler
import tensorflow as tf

ds = xr.open_dataset('/glade/derecho/scratch/sarahryu/mean_2000_2020.nc', engine='netcdf4')
ds = ds.isel(ens=slice(1,1494))
ds
```

```python
# Make target df

target_df = ds[['TWS_gridded_amean', 'EFLX_LH_TOT_gridded_amean']].drop_vars(['levgrnd', 'crashed']).to_dataframe().reset_index()
target_df["gridcell_ens"] = (
    target_df["gridcell"].astype(str) + "_" + target_df["ens"].astype(str)
)

target_df = target_df.set_index(["gridcell_ens"])
target_df = target_df.drop(columns=['ens', 'gridcell']) 

target_df
```

```python
# Forcings for each gridcell

vars = [
    name for name, da in ds.data_vars.items()
    if set(da.dims) == {"gridcell"}
]

forcings_df = ds[vars].to_dataframe()

forcings_df = forcings_df.drop(columns=['BSW','HKSAT', 'SUCSAT']) # these result in nans in the scaled dataset for some reason
forcings_df
```

```python
## Combine forcings with ppe params
# Keep gridcells next to each other so that they can be withheld from test set

ppe_params = pd.read_csv('/glade/derecho/scratch/linnia/leap-reu-mf-2025/ctsm6lhc_params.csv').drop(columns='member')

params = ppe_params.reset_index().merge(forcings_df.reset_index(), how="cross")
params["gridcell_ens"] = (
    params["gridcell"].astype(str) + "_" + params["index"].astype(str)
)

params = params.set_index(["gridcell_ens"])
params = params.drop(columns=['index', 'gridcell', 'levgrnd', 'grid1d_lat', 'grid1d_lon', 'grid1d_ixy', 'grid1d_jxy'])

# 71 because it include the soil properties I think
params
```

```python
# Train test split: withhold 20 ensemble members, withhold 10 grid cells

def train_test_set_indices(named_indices, list_1, list_2):
    """
    Inputs: named indices of df, lists of gridcells to put in test set, list of ens's to put in test set
    Output: numbered indices of train set, numbered indices of test set
    """
    # Initialize mask
    mask = pd.Series(False, index=named_indices)

    # Add indices to mask
    for i in list_1:
        mask = mask | named_indices.str.startswith(f'{i}_')
    for j in list_2:
        mask = mask | named_indices.str.endswith(f'_{j}')

    test = np.where(mask)[0]
    train = np.where(~mask)[0]
    return train, test


test_gridcells = random.sample(range(400), 10)
test_ens = random.sample(range(1, 1494), 20)
print(f"Test set: gridcells {test_gridcells}, ensemble members {test_ens}")

train_idx, test_idx = train_test_set_indices(target_df.index, test_gridcells, test_ens)
```

## MLP Emulator

```python
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras import regularizers

early_stop = EarlyStopping(
    monitor='val_loss',       # metric to monitor
    patience=10,              # epochs to wait for improvement
    restore_best_weights=True  # revert to best weights
)

var = 'EFLX_LH_TOT_gridded_amean'
y = target_df[var]
X = params
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]
X_train, X_test = X_scaled[train_idx], X_scaled[test_idx]

# Define MLP model
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(X.shape[1],)),  # Input layer
    tf.keras.layers.Dense(64, activation='relu'), # L1 regularization encourages sparse weights
    #tf.keras.layers.Dropout(0.2), # dropout
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(16, activation='relu'),
    tf.keras.layers.Dense(1)  # Output layer (single float)
])

# Compile model
model.compile(optimizer='adam', loss='mse', metrics=['mae'])

# Train model
history = model.fit(
    X_train, 
    y_train, 
    validation_split=0.1, 
    epochs=100, 
    batch_size=32, 
    callbacks=[early_stop],
    verbose=1
) 

# Evaluate on test set
test_loss, test_mae = model.evaluate(X_test, y_test)
print(f"\nTest MSE: {test_loss:.4f}, MAE: {test_mae:.4f}")

y_pred = model.predict(X_test).flatten()
r2 = r2_score(y_test,y_pred)
print(r2)

plt.figure(figsize=(10,6))
plt.plot([min(y_test),max(y_test)],[min(y_test),max(y_test)],c='k',linestyle='--',label='1:1 line')
plt.scatter(y_test,y_pred, s=5, alpha=0.5, label = f'r2={r2}')
plt.xlabel(f'CLM {var}')
plt.ylabel(f'Emulated {var}')
plt.legend()

plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from SALib.sample import saltelli
from SALib.analyze import sobol

# 1) Define your problem as before
names  = params.columns.tolist()
bounds = np.vstack([params.min(), params.max()]).T.tolist()
problem = {
    'num_vars': len(names),
    'names':    names,
    'bounds':   bounds
}

# 2) Sample *with* second‐order enabled
N = 1024
X = saltelli.sample(problem, N, calc_second_order=True)

# 3) Evaluate your MLP and flatten
Y = model.predict(scaler.transform(X)).ravel()

# 4) Analyze *with* second‐order indices
Si = sobol.analyze(
    problem, Y,
    calc_second_order=True,
    print_to_console=False
)

# 5) Extract S1 and ST
S1 = pd.Series(Si['S1'], index=names)
ST = pd.Series(Si['ST'], index=names)

# 6) Compute the “interaction‐only” portion
Int = ST - S1

# 7) Build a DataFrame and sort
df_plot = pd.DataFrame({
    'Main effect (S1)':   S1,
    'Interactions (ST−S1)': Int
}).sort_values('Main effect (S1)', ascending=False)

# 8) Plot stacked bars
ax = df_plot.plot(
    kind='bar',
    stacked=True,
    figsize=(10,8),
    color=['darkolivegreen','lightgrey']
)
ax.set_ylabel("Proportion of total variance")
ax.set_ylim(0, 1)
ax.legend(loc='upper right')
ax.set_title("Spatial Sensitivity of ET to Meteorology and PPE Parameters")

plt.tight_layout()
plt.show()
```

```python
var ='TWS_gridded_amean'
# log transform y

y = target_df[var]
y = np.clip(y, a_min=0, a_max=None) # zero all negatives (<0.1% of values)
X = params
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]
X_train, X_test = X_scaled[train_idx], X_scaled[test_idx]

y_train_log = np.log1p(y_train)
y_test_log = np.log1p(y_test)

# Define MLP model
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(X.shape[1],)),  # Input layer
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(1)  # Output layer (single float)
])

# Compile model
model.compile(optimizer='adam', loss='mse', metrics=['mae'])

# Train model
history = model.fit(X_train, y_train_log, validation_split=0.1, epochs=60, batch_size=32, verbose=1)

# Evaluate on test set
test_loss, test_mae = model.evaluate(X_test, y_test_log)
print(f"\nTest MSE: {test_loss:.4f}, MAE: {test_mae:.4f}")

y_pred_log = model.predict(X_test).flatten()
y_pred = np.expm1(y_pred_log)  # Inverse of log1p
r2 = r2_score(y_test, y_pred)
print(r2)

plt.figure(figsize=(10,6))
plt.plot([min(y_test),max(y_test)],[min(y_test),max(y_test)],c='k',linestyle='--',label='1:1 line')
plt.scatter(y_test,y_pred, s=5, alpha=0.5, label = f'r2={r2}')
plt.xlabel(f'CLM {var}')
plt.ylabel(f'Emulated {var}')
plt.legend()

plt.show()
```
